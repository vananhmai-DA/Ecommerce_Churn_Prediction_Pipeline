````python
risk_numeric_profile = (
    segmentation_df
    .groupby("risk_level")
    .agg(
        customers=("customer_id", "count"),
        churn_rate=("churn", "mean"),
        avg_churn_probability=("churn_probability", "mean"),
        avg_tenure=("tenure", "mean"),
        median_tenure=("tenure", "median"),
        complaint_rate=("complain", "mean"),
        has_complaint_rate=("has_complaint", "mean"),
        avg_days_since_last_order=("day_since_last_order", "mean"),
        avg_order_count=("order_count", "mean"),
        avg_coupon_used=("coupon_used", "mean"),
        avg_cashback_amount=("cashback_amount", "mean"),
        avg_satisfaction_score=("satisfaction_score", "mean"),
        avg_warehouse_to_home=("warehouse_to_home", "mean")
    )
    .reset_index()
)

risk_numeric_profile
````

````python
risk_category_profile = (
    segmentation_df
    .groupby(["risk_level", "preferred_order_cat"])
    .agg(customers=("customer_id", "count"))
    .reset_index()
)

risk_category_profile["share_within_risk_level"] = (
    risk_category_profile["customers"]
    / risk_category_profile.groupby("risk_level")["customers"].transform("sum")
)

risk_category_profile.sort_values(
    ["risk_level", "share_within_risk_level"],
    ascending=[True, False]
)
````

## Segmentation Insights

The prediction output was merged with the Gold customer table to analyze customer churn risk segments.

The model classified 853 customers as High Risk, representing 15.15% of the total customer base. Most customers were classified as Low Risk, while only a small share fell into the Medium Risk group.

The purpose of this segmentation is not only to label customers as High, Medium, or Low Risk, but also to understand how these groups differ and what actions the business should take for each group.

### High Risk Segment

High Risk customers show several clear data patterns.

First, this group has very low tenure. The average tenure of High Risk customers is around 2.63, compared with around 11.07 for Low Risk customers. The median tenure of the High Risk group is only 1. This suggests that many churn-risk customers are still in the early stage of their customer lifecycle.

Business implication:

Customers with short tenure may not have built strong loyalty or repeat-purchase habits yet. Therefore, the company should not wait until these customers become inactive. Early onboarding and engagement actions are important.

Recommended actions:

* Create onboarding campaigns for new or low-tenure customers.
* Send first-month or second-purchase incentives.
* Track whether new customers make another purchase within a short period.
* Provide personalized product recommendations after the first order.

Second, complaint behavior is strongly concentrated in the High Risk group. More than half of High Risk customers have submitted complaints. This suggests that service experience is an important churn driver.

Business implication:

For these customers, a generic discount may not be enough. If the churn signal comes from a complaint, the company should first fix the service issue before sending promotions.

Recommended actions:

* Prioritize High Risk customers with complaint history.
* Create a complaint recovery workflow.
* Follow up with affected customers after issue resolution.
* Offer compensation, cashback, or service recovery vouchers only after the complaint is addressed.

Third, High Risk customers are strongly concentrated in mobile-related categories. Mobile phone accounts for 38.34% of High Risk customers, and mobile accounts for 23.21%. Together, mobile-related categories represent 61.55% of the High Risk segment.

This is much higher than the Low Risk group, where mobile phone and mobile together represent around 32.20%.

Business implication:

Mobile-related purchases may be more one-off or replacement-driven, meaning customers may have fewer natural reasons to return after the first purchase. These categories may require a different retention strategy compared with repeat-driven categories such as grocery.

Recommended actions:

* Design mobile-category retention campaigns.
* Use accessory cross-sell campaigns after mobile purchases.
* Send warranty, care, or usage-related follow-up messages.
* Offer targeted cashback for second purchases in related categories.

Finally, High Risk customers also have lower average cashback amounts than Low Risk customers. This may suggest weaker incentive exposure or lower engagement with cashback-based retention mechanisms.

Business implication:

Targeted incentives may help increase engagement, especially when combined with low tenure and complaint signals.

Recommended actions:

* Use targeted cashback instead of broad discounts.
* Prioritize incentives for High Risk customers with low tenure, complaint history, or mobile-related purchases.
* Avoid giving large discounts to all customers equally.

### Medium Risk Segment

Medium Risk customers show mixed behavior. They are not as urgent as High Risk customers, but they already show some churn signals.

In terms of preferred order category, Medium Risk customers are spread across mobile phone, laptop & accessory, and mobile. This suggests that Medium Risk is a transition group rather than a single clearly defined customer type.

Business implication:

Medium Risk customers should be monitored before they become High Risk. The business should use lighter and lower-cost engagement actions for this group.

Recommended actions:

* Send light incentives such as coupons or free shipping.
* Monitor changes in days since last order, complaint status, and order frequency.
* Trigger stronger retention actions if the customer moves from Medium Risk to High Risk.
* Use personalized product recommendations instead of aggressive discounting.

### Low Risk Segment

Low Risk customers represent the majority of the customer base. This group is more concentrated in laptop & accessory, which accounts for 39.29% of Low Risk customers.

Compared with the High Risk group, Low Risk customers have higher tenure and lower complaint concentration. This suggests that they are relatively more stable and should not receive aggressive retention spending.

Business implication:

The company should maintain engagement with Low Risk customers but avoid unnecessary discounts.

Recommended actions:

* Maintain loyalty communication.
* Send regular product updates or personalized recommendations.
* Avoid over-discounting customers who are already stable.
* Continue monitoring for future changes in churn risk.

### Key Takeaway

The main insight from the risk segmentation is that High Risk customers are not just randomly selected by the model. They show clear behavioral patterns: lower tenure, higher complaint concentration, stronger concentration in mobile-related categories, and lower cashback amounts.

Therefore, retention actions should not be generic. The company should prioritize:

* Onboarding and early engagement for low-tenure customers.
* Complaint recovery for customers with service issues.
* Mobile-category cross-sell and second-purchase campaigns.
* Targeted cashback for selected High Risk customers.
